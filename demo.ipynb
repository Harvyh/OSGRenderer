```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from PyRenderer import Renderer
```

```python
viewport_size_x = 700
viewport_size_y = 700

azimuth = 0
elevation = 20
yaw = 0
distance_ratio = 1
field_of_view = 25

x = Renderer()
x.initialize(['mesh/Honda-Accord.3ds',
              'mesh/2012-VW-beetle-turbo.3ds'],
              viewport_size_x, viewport_size_y)
```

```python
x.setViewpoint(azimuth,elevation,yaw,distance_ratio,field_of_view)
rendering, depth = x.render()
rendering = rendering.transpose((2,1,0))
depth = depth.transpose((1,0))
```

```python
plt.imshow(rendering)
plt.show()
```

```python
plt.imshow(1-depth)
plt.show()
```

```python
x.setModelIndex(1)
x.setViewpoint(azimuth,elevation,yaw,distance_ratio,field_of_view)
rendering, depth = x.render()
rendering = rendering.transpose((2,1,0))
depth = depth.transpose((1,0))

plt.imshow(rendering)
plt.show()
```

```python
plt.imshow(1-depth)
plt.show()
```
